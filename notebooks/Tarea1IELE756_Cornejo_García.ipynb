### 0.2 Filter to selected comunas

The dataset was filtered to the assigned comunas: Renca, Lo Barnechea, and Buin. Since the dataset uses a numeric comuna code, the corresponding codes from the Census dictionary were used.
First, the `vivienda` table was filtered using these comuna codes. Then, the filtered set of viviendas was used to subset the `hogar` and `persona` tables through the common key `id_vivienda`. This ensures consistency across all three tables and keeps only the observations belonging to the selected comunas.

Finally, the number of rows in each filtered table was reported to understand the size of the resulting datasets.

```python
MY_COMUNAS = [13128, 13115, 13402]  # Codes for Renca, Lo Barnechea, Buin
# Filter the vivienda table to the selected comunas
viv = vivienda[vivienda["comuna"].isin(MY_COMUNAS)]

# Filter hogar table, keep only hogares belonging to the filtered viviendas
hog = hogar[hogar["id_vivienda"].isin(viv["id_vivienda"])]

# Filter the persona table to keep only people living in the selected viviendas
per = persona[persona["id_vivienda"].isin(viv["id_vivienda"])]

# Report the size of each filtered table
print("Vivienda:", viv.shape)
print("Hogar:", hog.shape)
print("Persona:", per.shape)
```

### 0.3 Join the tables

The `persona` table was merged with the `hogar` table using the keys `id_vivienda` and `id_hogar`, matching each individual with their corresponding household information.
Then, the resulting table was merged with the `vivienda` table using `id_vivienda`.
A left join was used in both merges to ensure that all observations from the `persona` table are preserved, even if some information is missing in the other tables.
Finally, it was verified that the number of rows in the final dataset matches the number of rows in the `persona` table, confirming that no rows were lost or duplicated during the join process.

```python
# Merge household information into the persona dataset using id_vivienda and id_hogar
# Then merge dwelling information using id_vivienda
df = per.merge(hog, on=["id_vivienda", "id_hogar"], how="left") \
        .merge(viv, on="id_vivienda", how="left")
# Check that no rows were lost or duplicated
assert len(df) == len(per), "Row count mismatch after join!"
# Inspect the result
df.shape
df.info()
```

The resulting dataset contains 373,211 observations, which matches the number of rows in the `persona` table. This confirms that the join was performed correctly without losing or duplicating records.
The final number of columns (28) is not equal to the simple sum of the columns in the three tables (20 + 5 + 6) because some variables are shared across tables and are used as merge keys. Merge keys such as `id_vivienda` and `id_hogar` are not duplicated in the final dataset.

## Part 1: Demographic Profile

### 1.1 Age pyramid by sex

A population pyramid was created for the three selected comunas combined using 5-year age groups. Observations with `edad = -99` were removed because this value represents missing information. In addition, observations with `p25_lug_nacimiento_rec = -99` were also excluded, since they correspond to missing responses.

Age groups were created using `pd.cut()` with intervals from 0–4 up to 80+.

Males were plotted on the left using negative values, and females on the right using positive values. The pyramid distinguishes between Chilean-born and foreign-born individuals using `p25_lug_nacimiento_rec`, where `1` indicates individuals born in Chile and `2` indicates individuals born outside Chile. This allows for a clearer visualization of the age structure of the immigrant population.

```python
# Import plotting library
import matplotlib.pyplot as plt

# Keep only valid observations (not nulls)
df_age = df[(df["edad"] != -99) & (df["p25_lug_nacimiento_rec"] != -99)].copy()

# Create 5-year age groups
bins = list(range(0, 81, 5)) + [200]
labels = [f"{i}-{i+4}" for i in range(0, 80, 5)] + ["80+"]

df_age["age_group"] = pd.cut(df_age["edad"], bins=bins, labels=labels, right=False)

# Define Chilean-born vs Foreign-born. Code 1 = born in Chile, code 2 = born outside Chile
df_age["nativity"] = df_age["p25_lug_nacimiento_rec"].map({
    1: "Chilean-born",
    2: "Foreign-born"
})
# Count population by age group, sex, and nativity
pyramid_counts = (
    df_age.groupby(["age_group", "sexo", "nativity"], observed=False)
    .size()
    .reset_index(name="count")
)

# Create separate series for each group and keep all age groups in the correct order
male_chilean = (
    pyramid_counts[
        (pyramid_counts["sexo"] == 1) &
        (pyramid_counts["nativity"] == "Chilean-born")
    ]
    .set_index("age_group")["count"]
    .reindex(labels, fill_value=0)
)

female_chilean = (
    pyramid_counts[
        (pyramid_counts["sexo"] == 2) &
        (pyramid_counts["nativity"] == "Chilean-born")
    ]
    .set_index("age_group")["count"]
    .reindex(labels, fill_value=0)
)

male_foreign = (
    pyramid_counts[
        (pyramid_counts["sexo"] == 1) &
        (pyramid_counts["nativity"] == "Foreign-born")
    ]
    .set_index("age_group")["count"]
    .reindex(labels, fill_value=0)
)

female_foreign = (
    pyramid_counts[
        (pyramid_counts["sexo"] == 2) &
        (pyramid_counts["nativity"] == "Foreign-born")
    ]
    .set_index("age_group")["count"]
    .reindex(labels, fill_value=0)
)

# Plot the population pyramid
plt.figure(figsize=(11, 8))

plt.barh(labels, -male_chilean, label="Male - Chilean-born", alpha=0.8)
plt.barh(labels, -male_foreign, label="Male - Foreign-born", alpha=0.8)

plt.barh(labels, female_chilean, label="Female - Chilean-born", alpha=0.8)
plt.barh(labels, female_foreign, label="Female - Foreign-born", alpha=0.8)

# Add labels and title
plt.xlabel("Population")
plt.ylabel("Age group")
plt.title("Population Pyramid by Sex and Nativity")

# Add a vertical line at zero to separate both sides
plt.axvline(0, linewidth=1)

# Improve layout and legend
plt.legend()
plt.tight_layout()
plt.show()
```

The population pyramid shows that Chilean-born residents represents most of the population across all age groups. In contrast, the foreign-born population appears more concentrated in working-age groups, with a smaller presence among children and older adults.

### 1.2 Dependency ratio

The age dependency ratio was computed for each selected comuna using the following formula:

Dependency ratio = (population aged 0-14 + population aged 65+) /
population aged 15-64

Following the assignment instructions, observations with missing age (`edad = -99`) were excluded. Ages were grouped into three categories: 0-14, 15-64, and 65+. The result is presented as a small table with one row per comuna.

```python
# Keep valid ages (not nulls)
df_dep = df[df["edad"] != -99].copy()

# Age groups
df_dep["age_group_simple"] = pd.cut(
    df_dep["edad"],
    bins=[0, 15, 65, 200],
    labels=["0-14", "15-64", "65+"],
    right=False
)

# Count by comuna and age group
dep_table = df_dep.groupby(["comuna", "age_group_simple"], observed=False).size().unstack(fill_value=0)

# Dependency ratio
dep_table["dependency_ratio"] = (
    dep_table["0-14"] + dep_table["65+"]
) / dep_table["15-64"]

# Reset index
dep_table = dep_table.reset_index()

# Map comuna codes to names
comuna_map = {
    13128: "Renca",
    13115: "Lo Barnechea",
    13402: "Buin"
}

dep_table["comuna"] = dep_table["comuna"].map(comuna_map)

# Keep only final columns
dep_table = dep_table[["comuna", "dependency_ratio"]]

# Remove the extra header name
dep_table.columns.name = None

dep_table
```

The dependency ratio measures how many people depend on the working-age population. It is calculated as the number of individuals aged 0–14 and 65+ divided by those aged 15–64.

The dependency ratios are quite similar across the three comunas, all close to 0.5. This indicates that for every 100 working-age individuals, there are about 50 dependents.

Overall, this value appears moderate and suggests that the working-age population is larger than the dependent population. The similarity across comunas also points to a fairly homogeneous demographic structure.


### 1.3 Household size distribution

Household sizes were grouped into categories from 1 to 7, with households of 8 or more members grouped into a single category labeled "8+".

The distribution is presented separately for Chilean-born and foreign-born households.

Households were classified by nationality using the head of household (identified by `parentesco = 1`). The nativity of the household head was used to determine whether the household is Chilean-born or foreign-born, based on the variable `p25_lug_nacimiento_rec`.

```python
import matplotlib.pyplot as plt

# Compute household size
# Each row in df represents one person, so counting rows per household gives household size
household_size = (
    df.groupby(["id_vivienda", "id_hogar"])
      .size()
      .reset_index(name="household_size")
)

# Keep only the head of household
# parentesco == 1 identifies the head of household
household_head = df.loc[
    df["parentesco"] == 1,
    ["id_vivienda", "id_hogar", "p25_lug_nacimiento_rec"]
].copy()

# Remove invalid birthplace values coded as -99
household_head = household_head[household_head["p25_lug_nacimiento_rec"] != -99]

# Keep only one head per household
household_head = household_head.drop_duplicates(subset=["id_vivienda", "id_hogar"])

# Merge household size with the nativity of the head of household
household_data = household_size.merge(
    household_head,
    on=["id_vivienda", "id_hogar"],
    how="inner"
)

# Define nationality group based on the recoded birthplace of the head of household
# 1 = Chilean-born, 2 = Foreign-born
household_data["nationality_group"] = household_data["p25_lug_nacimiento_rec"].map({
    1: "Chilean-born",
    2: "Foreign-born"
})

# Group household sizes above 8 into one category
household_data["size_group"] = household_data["household_size"].apply(
    lambda x: "8+" if x >= 8 else str(x)
)

# Compute the household size distribution
distribution = (
    household_data.groupby(["size_group", "nationality_group"], observed=False)
    .size()
    .unstack(fill_value=0)
)

# Reorder the categories on the x-axis
size_order = [str(i) for i in range(1, 8)] + ["8+"]
distribution = distribution.reindex(size_order, fill_value=0)

# Plot the bar chart
distribution.plot(kind="bar", figsize=(8, 6))

plt.xlabel("Household size")
plt.ylabel("Count")
plt.title("Household Size Distribution by Nativity of Household Head")
plt.legend(title="Group")
plt.tight_layout()
plt.show()
```

In both Chilean-born and foreign-born households, sizes between 2 and 4 members appear most frequently.

### 1.4 Education and employment

For each selected comuna, two indicators were computed separately for Chilean-born and foreign-born individuals.

First, the mean years of schooling (`escolaridad`) was calculated for individuals aged 25 and older. Observations with missing values (`-99`) were excluded.

Second, the employment rate was computed for individuals aged 15–64. This was defined as the share of individuals whose labor force status (`sit_fuerza_trabajo`) corresponds to being employed (code 1), according to the data dictionary.

The results are presented as grouped bar charts, where each comuna is shown with separate bars for Chilean-born and foreign-born populations.

```python
# 1. Mean years of schooling (age 25+)
# Keep valid observations
df_edu = df[
    (df["edad"] >= 25) &
    (df["edad"] != -99) &
    (df["escolaridad"] != -99) &
    (df["p25_lug_nacimiento_rec"] != -99)
].copy()

# Define nativity group
df_edu["nativity"] = df_edu["p25_lug_nacimiento_rec"].map({
    1: "Chilean-born",
    2: "Foreign-born"
})

# Compute mean schooling
edu_table = (
    df_edu.groupby(["comuna", "nativity"])["escolaridad"]
    .mean()
    .reset_index(name="mean_schooling")
)

# 2. Employment rate (age 15–64)
# Keep valid observations
df_emp = df[
    (df["edad"] >= 15) &
    (df["edad"] < 65) &
    (df["edad"] != -99) &
    (df["sit_fuerza_trabajo"] != -99) &
    (df["p25_lug_nacimiento_rec"] != -99)
].copy()

# Define nativity group
df_emp["nativity"] = df_emp["p25_lug_nacimiento_rec"].map({
    1: "Chilean-born",
    2: "Foreign-born"
})

# Define employment status (1 = employed)
df_emp["employed"] = (df_emp["sit_fuerza_trabajo"] == 1).astype(int)

# Compute employment rate
emp_table = (
    df_emp.groupby(["comuna", "nativity"])["employed"]
    .mean()
    .reset_index(name="employment_rate")
)

# 3. Merge results
final_table = edu_table.merge(
    emp_table,
    on=["comuna", "nativity"],
    how="inner"
)

# Map comuna names
comuna_map = {
    13128: "Renca",
    13115: "Lo Barnechea",
    13402: "Buin"
}

final_table["comuna_name"] = final_table["comuna"].map(comuna_map)

# 4. Plot grouped bar charts
# Education chart
edu_pivot = final_table.pivot(
    index="comuna_name",
    columns="nativity",
    values="mean_schooling"
)

edu_pivot.plot(kind="bar", figsize=(8, 5))
plt.title("Mean Years of Schooling (Age 25+)")
plt.ylabel("Years")
plt.xlabel("Comuna")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Employment chart
emp_pivot = final_table.pivot(
    index="comuna_name",
    columns="nativity",
    values="employment_rate"
)

emp_pivot.plot(kind="bar", figsize=(8, 5))
plt.title("Employment Rate (Age 15–64)")
plt.ylabel("Rate")
plt.xlabel("Comuna")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Show final table
final_table
```

Foreign-born individuals show higher employment rates across all comunas. This difference is mainly explained by their lower share of individuals outside the labor force, compared to Chilean-born populations.

In terms of education, foreign-born individuals also exhibit slightly higher average years of schooling, particularly in Renca and Buin.

### 2.1 Percentage foreign-born by comuna

The share of foreign-born individuals was computed for each comuna using the recoded variable `p25_lug_nacimiento_rec`, where code 1 indicates Chilean-born and code 2 indicates foreign-born.

Observations with missing values (`-99`) were excluded. The foreign-born share is defined as the proportion of individuals classified as foreign-born over the total population in each comuna.

```python
# Keep only observations with valid nativity information
df_mig = df[df["p25_lug_nacimiento_rec"] != -99].copy()

# Define foreign-born indicator
# 1 = Chilean-born, 2 = Foreign-born
df_mig["foreign_born"] = (df_mig["p25_lug_nacimiento_rec"] == 2).astype(int)

# Compute the proportion of foreign-born population by comuna
foreign_share = (
    df_mig.groupby("comuna")["foreign_born"]
    .mean()
    .reset_index(name="foreign_share")
)

# Convert from proportion to percentage
foreign_share["foreign_percentage"] = foreign_share["foreign_share"] * 100

# Map comuna codes to comuna names
comuna_map = {
    13128: "Renca",
    13115: "Lo Barnechea",
    13402: "Buin"
}

foreign_share["comuna_name"] = foreign_share["comuna"].map(comuna_map)

# Keep final columns
foreign_share = foreign_share[["comuna", "comuna_name", "foreign_share", "foreign_percentage"]]

# Sort rows
foreign_share = foreign_share.sort_values("comuna_name").reset_index(drop=True)

foreign_share
```

### 2.4 Arrival period of immigrants

Using the variable `p26_llegada_periodo`, the arrival period of foreign-born residents was analyzed.

Only individuals classified as foreign-born (`p25_lug_nacimiento_rec == 2`) were included. Observations with missing values (`-99`) and not applicable values (`NA`) were excluded.

Since the variable is coded numerically, it was mapped into descriptive arrival periods based on the official variable dictionary.

The code performs the following steps:

- Filters the dataset to include only foreign-born individuals with valid arrival information  
- Maps numeric codes into descriptive time intervals  
- Computes the distribution of arrival periods  
- Visualizes the results as a bar chart ordered chronologically  

The results are presented as a bar chart.

```python
# Keep only foreign-born individuals with valid arrival period
df_arrival = df[
    (df["p25_lug_nacimiento_rec"] == 2) &
    (~df["p26_llegada_periodo"].isin([-99]))
].copy()

# Mapping based on official dictionary
arrival_map = {
    1: "2023–2024",
    2: "2020–2022",
    3: "2017–2019",
    4: "2014–2016",
    5: "2010–2013",
    6: "2000–2009",
    7: "1990–1999",
    8: "Before 1990"
}

df_arrival["arrival_period"] = df_arrival["p26_llegada_periodo"].map(arrival_map)

# Remove any remaining NA after mapping
df_arrival = df_arrival[df_arrival["arrival_period"].notna()]

# Count values (ordered for correct visualization)
order = [
    "Before 1990",
    "1990–1999",
    "2000–2009",
    "2010–2013",
    "2014–2016",
    "2017–2019",
    "2020–2022",
    "2023–2024"
]

arrival_counts = (
    df_arrival["arrival_period"]
    .value_counts()
    .reindex(order)
)

# Plot
plt.figure(figsize=(10, 6))
arrival_counts.plot(kind="bar")

plt.title("Arrival Period of Immigrants")
plt.xlabel("Arrival period")
plt.ylabel("Number of individuals")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

arrival_counts
```

The total population was computed from the joined census table by counting person-level observations within each comuna. This information was then merged with the shapefile using the comuna code.

The map includes a title, a legend, and comuna labels.

```python
import geopandas as gpd
# 1. Load the shapefile
comunas_gdf = gpd.read_file("comunas.shp")
# 2. Compute total population by comuna
# Since df is at the person level, counting rows gives total population
pop_by_comuna = (
    df.groupby("comuna")
    .size()
    .reset_index(name="total_population")
)

# 3. Merge census data with the shapefile
map_pop = comunas_gdf.merge(
    pop_by_comuna,
    left_on="cod_comuna",
    right_on="comuna",
    how="inner"
)

# 4. Plot choropleth map
fig, ax = plt.subplots(figsize=(10, 10))

map_pop.plot(
    column="total_population",
    cmap="Blues",
    linewidth=0.8,
    edgecolor="black",
    legend=True,
    ax=ax
)

# Add title and remove axes
ax.set_title("Total Population by Comuna", fontsize=14)
ax.axis("off")

# Add comuna labels if legible
for _, row in map_pop.iterrows():
    centroid = row.geometry.centroid
    ax.text(
        centroid.x,
        centroid.y,
        row["Comuna"],
        fontsize=8,
        ha="center"
    )

plt.tight_layout()
plt.show()
```

The map shows the spatial distribution of total population across the three selected comunas.

Renca appears as the most populated comuna, followed by Buin, while Lo Barnechea has a relatively lower population. This visualization allows for a direct spatial comparison of population size across the selected comunas.

### 3.2 Choropleth map: % foreign born by comuna

A second choropleth map was created to show the percentage of foreign-born residents in each selected comuna.

The percentage was calculated using the recoded birthplace variable `p25_lug_nacimiento_rec`, where code `2` identifies foreign-born individuals. Observations with missing values (`-99`) were excluded.

The resulting comuna-level percentages were merged with the comuna shapefile and displayed using a sequential color palette.

```python
# 1. Load the shapefile
comunas_gdf = gpd.read_file("comunas.shp")

# 2. Compute the percentage of foreign-born residents by comuna
# Keep only observations with valid birthplace information
df_foreign = df[df["p25_lug_nacimiento_rec"] != -99].copy()

# Create a foreign-born indicator
# Code 2 in p25_lug_nacimiento_rec corresponds to foreign-born individuals
df_foreign["foreign_born"] = (df_foreign["p25_lug_nacimiento_rec"] == 2).astype(int)

# Compute the percentage of foreign-born residents by comuna
foreign_pct = (
    df_foreign.groupby("comuna")["foreign_born"]
    .mean()
    .reset_index(name="foreign_share")
)

# Convert the proportion into percentage
foreign_pct["foreign_percentage"] = foreign_pct["foreign_share"] * 100

# 3. Merge percentage data with the shapefile
# The shapefile uses 'cod_comuna' and the census table uses 'comuna'
map_foreign = comunas_gdf.merge(
    foreign_pct,
    left_on="cod_comuna",
    right_on="comuna",
    how="inner"
)

# 4. Keep only the three selected comunas
selected_comunas = [13128, 13115, 13402]
map_foreign = map_foreign[map_foreign["cod_comuna"].isin(selected_comunas)].copy()

# 5. Plot choropleth map
fig, ax = plt.subplots(figsize=(8, 8))

map_foreign.plot(
    column="foreign_percentage",
    cmap="YlOrRd",
    linewidth=1,
    edgecolor="black",
    legend=True,
    ax=ax
)

# Add title and remove axes
ax.set_title("Percentage of Foreign-born Residents by Comuna", fontsize=14)
ax.axis("off")

# Add comuna labels
for _, row in map_foreign.iterrows():
    centroid = row.geometry.centroid
    ax.text(
        centroid.x,
        centroid.y,
        row["Comuna"],
        fontsize=9,
        ha="center",
        weight="bold"
    )

plt.tight_layout()
plt.show()
```

The map shows clear spatial differences in the share of foreign-born residents across the three comunas.

Lo Barnechea has the highest percentage of foreign-born residents, followed by Renca, while Buin has the lowest share. 


## Part 4: Comuna-Level Summary Table

Captura de pantalla 2026-03-26 011825.png

A comuna-level summary table was constructed as the main output of the assignment.

The table combines population, migration, age structure, education, employment, and dependency indicators for each selected comuna. Indicators were computed separately for Chilean-born and foreign-born populations where required.

Missing values coded as `-99` were excluded from calculations rather than treated as valid observations. The final table is displayed in the notebook and saved as a CSV file for future use.

```python
import os

def dep_ratio(data_frame):
    # Keep only valid ages
    df_dep_temp = data_frame[data_frame["edad"] != -99].copy()

    # Create age groups for dependency ratio
    df_dep_temp["age_group_simple"] = pd.cut(
        df_dep_temp["edad"],
        bins=[0, 15, 65, 200],
        labels=["0-14", "15-64", "65+"],
        right=False
    )

    # Count population by age group
    dep_counts = df_dep_temp.groupby("age_group_simple", observed=False).size()

    # Ensure all age groups are present
    dep_counts = dep_counts.reindex(["0-14", "15-64", "65+"], fill_value=0)

    # Compute dependency ratio
    dependency_ratio_val = (
        dep_counts["0-14"] + dep_counts["65+"]
    ) / dep_counts["15-64"] if dep_counts["15-64"] > 0 else 0

    return dependency_ratio_val


def build_summary_row(group):
    """Compute all required comuna-level indicators."""

    res = {}

    # Population counts
    nativity_valid = group[group["p25_lug_nacimiento_rec"].isin([1, 2])].copy()
    nativity_valid["nativity"] = nativity_valid["p25_lug_nacimiento_rec"].map({
        1: "CH",
        2: "EX"
    })

    res["codigo_comuna"] = group.name
    res["pop_total"] = len(group)
    res["pop_chilean"] = len(nativity_valid[nativity_valid["nativity"] == "CH"])
    res["pop_foreign"] = len(nativity_valid[nativity_valid["nativity"] == "EX"])
    res["pct_foreign"] = (
        res["pop_foreign"] / res["pop_total"] * 100
        if res["pop_total"] > 0 else 0
    )

    # Median age
    age_valid = nativity_valid[nativity_valid["edad"] != -99].copy()

    res["median_age_chilean"] = age_valid.loc[
        age_valid["nativity"] == "CH", "edad"
    ].median()

    res["median_age_foreign"] = age_valid.loc[
        age_valid["nativity"] == "EX", "edad"
    ].median()

    # Mean schooling (age 25+)
    schooling_valid = nativity_valid[
        (nativity_valid["edad"] >= 25) &
        (nativity_valid["edad"] != -99) &
        (nativity_valid["escolaridad"] != -99)
    ].copy()

    res["mean_schooling_chilean"] = schooling_valid.loc[
        schooling_valid["nativity"] == "CH", "escolaridad"
    ].mean()

    res["mean_schooling_foreign"] = schooling_valid.loc[
        schooling_valid["nativity"] == "EX", "escolaridad"
    ].mean()

    # Employment rate (age 15-64)
    emp_valid = nativity_valid[
        (nativity_valid["edad"] >= 15) &
        (nativity_valid["edad"] <= 64) &
        (nativity_valid["edad"] != -99) &
        (nativity_valid["sit_fuerza_trabajo"] != -99)
    ].copy()

    emp_ch = emp_valid[emp_valid["nativity"] == "CH"]
    emp_ex = emp_valid[emp_valid["nativity"] == "EX"]

    res["emp_rate_chilean"] = (
        (emp_ch["sit_fuerza_trabajo"] == 1).mean() if len(emp_ch) > 0 else 0
    )

    res["emp_rate_foreign"] = (
        (emp_ex["sit_fuerza_trabajo"] == 1).mean() if len(emp_ex) > 0 else 0
    )

    # Dependency ratio
    res["dependency_ratio"] = dep_ratio(group)

    return pd.Series(res)


# Map comuna codes to comuna names
comuna_names = {
    13128: "Renca",
    13115: "Lo Barnechea",
    13402: "Buin"
}

# Build summary table
summary = df.groupby("comuna").apply(build_summary_row).reset_index(drop=True)

# Add comuna names
summary["nombre_comuna"] = summary["codigo_comuna"].map(comuna_names)

# Reorder columns to match the assignment
summary = summary[
    [
        "codigo_comuna",
        "nombre_comuna",
        "pop_total",
        "pop_chilean",
        "pop_foreign",
        "pct_foreign",
        "median_age_chilean",
        "median_age_foreign",
        "mean_schooling_chilean",
        "mean_schooling_foreign",
        "emp_rate_chilean",
        "emp_rate_foreign",
        "dependency_ratio"
    ]
]

# Round numeric columns
summary = summary.round(2)

# Display final summary table
print("FINAL SUMMARY TABLE - TASK 1")
display(summary)

# Save as CSV
os.makedirs("output", exist_ok=True)
summary.to_csv("output/tarea1_comuna_summary.csv", index=False)
```

The summary table highlights clear differences across the three comunas in terms of population size, migration patterns, education, and labor market outcomes.

Renca has the largest total population, while Lo Barnechea shows the highest share of foreign-born residents (10.83%), followed by Renca (8.72%) and Buin (5.10%). This confirms that migration is unevenly distributed across comunas.

In terms of age structure, foreign-born individuals tend to be slightly older in Lo Barnechea but similar or younger in the other comunas. Overall, median ages are relatively close across groups.

Educational attainment is consistently higher among foreign-born residents in all three comunas, with noticeable gaps especially in Renca and Buin. Employment rates are also higher for foreign-born individuals in every comuna, suggesting strong labor market participation among migrants.

Finally, dependency ratios are very similar across comunas (around 0.46–0.47), indicating comparable demographic structures.

Overall, the results show that while population size and migration intensity differ across comunas, foreign-born residents tend to exhibit higher education levels and employment rates across all areas.
